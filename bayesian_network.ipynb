```python
import pandas as pd

# Dataset
data = pd.DataFrame({
    'Rain': [0,0,1,1,0,1,1,0],
    'TrafficJam': [1,1,1,0,1,0,0,0],
    'ArriveLate': [1,1,1,0,0,1,1,1]
})

print("Dataset:\n", data)

# Step 1: P(Rain)
p_rain = data['Rain'].value_counts(normalize=True)
print("\nP(Rain):\n", p_rain)

# Step 2: P(TrafficJam | Rain)
p_traffic_given_rain = pd.crosstab(data['TrafficJam'], data['Rain'], normalize='columns')
print("\nP(TrafficJam | Rain):\n", p_traffic_given_rain)

# Step 3: P(ArriveLate | TrafficJam)
p_late_given_traffic = pd.crosstab(data['ArriveLate'], data['TrafficJam'], normalize='columns')
print("\nP(ArriveLate | TrafficJam):\n", p_late_given_traffic)

# Step 4: Query → P(ArriveLate | Rain = 1)

# Use total probability:
# P(Late | Rain) = Σ P(Late | TrafficJam) * P(TrafficJam | Rain)

p_late_given_rain = (
    p_late_given_traffic[1][1] * p_traffic_given_rain[1][1] +   # Traffic=1
    p_late_given_traffic[1][0] * p_traffic_given_rain[0][1]     # Traffic=0
)

p_not_late_given_rain = (
    p_late_given_traffic[0][1] * p_traffic_given_rain[1][1] +
    p_late_given_traffic[0][0] * p_traffic_given_rain[0][1]
)

print("\nFinal Result:")
print("P(ArriveLate=1 | Rain=1):", round(p_late_given_rain, 3))
print("P(ArriveLate=0 | Rain=1):", round(p_not_late_given_rain, 3))
```

```python
import pandas as pd

# Dataset
data = pd.DataFrame({
    'StudyHours': [2,5,3,6,1,4,5,2],
    'Difficulty': [1,2,2,1,2,1,2,1],  # 1=Easy, 2=Hard
    'Pass':       [0,1,1,1,0,1,0,1]
})

# Convert StudyHours → binary
data['Study'] = data['StudyHours'].apply(lambda x: 1 if x >= 4 else 0)

print("Dataset:\n", data)

# Step 1: P(Pass)
p_pass = data['Pass'].value_counts(normalize=True)

# Step 2: Conditional probabilities
p_study_given_pass = pd.crosstab(data['Study'], data['Pass'], normalize='columns')
p_diff_given_pass = pd.crosstab(data['Difficulty'], data['Pass'], normalize='columns')

print("\nP(Study | Pass):\n", p_study_given_pass)
print("\nP(Difficulty | Pass):\n", p_diff_given_pass)

# SAFE ACCESS using .loc[row, column]

# Query: Study=0 (Low), Difficulty=2 (Hard)

num = (
    p_pass[1] *
    p_study_given_pass.loc[0, 1] *
    p_diff_given_pass.loc[2, 1]
)

den = num + (
    p_pass[0] *
    p_study_given_pass.loc[0, 0] *
    p_diff_given_pass.loc[2, 0]
)

print("\nFinal Result:")
print("P(Pass=1 | LowStudy, Hard):", round(num/den, 3))
```

```python
import pandas as pd

data = pd.DataFrame({
    'Discount': [1,0,1,0,1,0,1,0],   # 1=High
    'Quality':  [1,0,1,1,0,0,0,1],   # 1=Good
    'Buy':      [1,0,1,1,0,0,0,1]
})

# P(Buy)
p_buy = data['Buy'].value_counts(normalize=True)

# P(Discount | Buy)
p_disc_given_buy = pd.crosstab(data['Discount'], data['Buy'], normalize='columns')

# P(Quality | Buy)
p_qual_given_buy = pd.crosstab(data['Quality'], data['Buy'], normalize='columns')

# Query: Discount=0, Quality=1
num = p_buy[1] * p_disc_given_buy[0][1] * p_qual_given_buy[1][1]
den = num + (p_buy[0] * p_disc_given_buy[0][0] * p_qual_given_buy[1][0])

print("P(Buy=1 | LowDiscount, GoodQuality):", round(num/den, 3))
```

```python
import pandas as pd

data = pd.DataFrame({
    'Battery': [1,0,1,0,1,0,1,0],
    'Fuel':    [1,0,0,1,1,0,0,1],
    'Start':   [1,0,0,0,1,0,0,0]
})

# P(Start)
p_start = data['Start'].value_counts(normalize=True)

# P(Battery | Start)
p_batt_given_start = pd.crosstab(data['Battery'], data['Start'], normalize='columns')

# P(Fuel | Start)
p_fuel_given_start = pd.crosstab(data['Fuel'], data['Start'], normalize='columns')

# Query: Battery=0, Fuel=1
num = p_start[1] * p_batt_given_start[0][1] * p_fuel_given_start[1][1]
den = num + (p_start[0] * p_batt_given_start[0][0] * p_fuel_given_start[1][0])

print("P(Start=1 | BatteryBad, FuelFull):", round(num/den, 3))
```
