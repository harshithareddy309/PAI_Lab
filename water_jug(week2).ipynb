```python
from collections import deque

jug1 = int(input("Enter capacity of Jug1: "))
jug2 = int(input("Enter capacity of Jug2: "))
target = int(input("Enter target amount: "))

def bfs_steps():
    # state: (amt1, amt2)
    # each queue entry: (amt1, amt2, path)
    q = deque()
    q.append((0, 0, []))
    visited = set()

    while q:
        a, b, path = q.popleft()

        if (a, b) in visited:
            continue
        visited.add((a, b))

        # Check goal
        if a == target or b == target:
            for i, step in enumerate(path, 1):
                print(f"Step {i}: {step}")
            print("\nFinal:", (a, b))
            return

        # All possible moves
        moves = []

        # Fill jug1
        if a != jug1:
            moves.append((jug1, b, f"Fill Jug1 -> ({jug1},{b})"))

        # Fill jug2
        if b != jug2:
            moves.append((a, jug2, f"Fill Jug2 -> ({a},{jug2})"))

        # Empty jug1
        if a != 0:
            moves.append((0, b, f"Empty Jug1 -> (0,{b})"))

        # Empty jug2
        if b != 0:
            moves.append((a, 0, f"Empty Jug2 -> ({a},0)"))

        # Pour jug1 -> jug2
        transfer = min(a, jug2 - b)
        if transfer > 0:
            newA = a - transfer
            newB = b + transfer
            moves.append((newA, newB,
                          f"Pour Jug1->Jug2 ({a},{b}) -> ({newA},{newB})"))

        # Pour jug2 -> jug1
        transfer = min(b, jug1 - a)
        if transfer > 0:
            newA = a + transfer
            newB = b - transfer
            moves.append((newA, newB,
                          f"Pour Jug2->Jug1 ({a},{b}) -> ({newA},{newB})"))

        # Push moves into queue
        for newA, newB, action in moves:
            q.append((newA, newB, path + [action]))

    print("No solution")

print("\nDetailed Steps:")
bfs_steps()
```

```python
def dfs_steps(jug1, jug2, target):
    visited = set()

    def dfs(a, b, path):
        # If goal reached
        if a == target or b == target:
            for i, step in enumerate(path, 1):
                print(f"Step {i}: {step}")
            print("\nFinal:", (a, b))
            return True

        # Mark visited
        visited.add((a, b))

        # List all valid moves with action descriptions
        moves = []

        # Fill Jug1
        if a != jug1:
            moves.append((jug1, b, f"Fill Jug1 -> ({jug1},{b})"))

        # Fill Jug2
        if b != jug2:
            moves.append((a, jug2, f"Fill Jug2 -> ({a},{jug2})"))

        # Empty Jug1
        if a != 0:
            moves.append((0, b, f"Empty Jug1 -> (0,{b})"))

        # Empty Jug2
        if b != 0:
            moves.append((a, 0, f"Empty Jug2 -> ({a},0)"))

        # Pour Jug1 -> Jug2
        transfer = min(a, jug2 - b)
        if transfer > 0:
            newA = a - transfer
            newB = b + transfer
            moves.append((newA, newB,
                          f"Pour Jug1->Jug2 ({a},{b}) -> ({newA},{newB})"))

        # Pour Jug2 -> Jug1
        transfer = min(b, jug1 - a)
        if transfer > 0:
            newA = a + transfer
            newB = b - transfer
            moves.append((newA, newB,
                          f"Pour Jug2->Jug1 ({a},{b}) -> ({newA},{newB})"))

        # Try all moves
        for newA, newB, action in moves:
            if (newA, newB) not in visited:
                if dfs(newA, newB, path + [action]):
                    return True

        return False

    if not dfs(0, 0, []):
        print("No solution")

# Getting user input and running
jug1 = int(input("Enter capacity of Jug1: "))
jug2 = int(input("Enter capacity of Jug2: "))
target = int(input("Enter target amount: "))

print("\nDFS Detailed Steps:")
dfs_steps(jug1, jug2, target)
```

```python
from collections import deque

jug1 = int(input("Enter capacity of Jug1: "))
jug2 = int(input("Enter capacity of Jug2: "))
target = int(input("Enter target amount: "))

def bfs_with_memo():
    # queue holds (amt1, amt2, path_of_actions)
    q = deque()
    q.append((0, 0, []))

    # visited (memo) stores already seen states
    visited = set()

    while q:
        a, b, path = q.popleft()

        # skip if already visited
        if (a, b) in visited:
            continue
        visited.add((a, b))

        # check target
        if a == target or b == target:
            for i, step in enumerate(path, 1):
                print(f"Step {i}: {step}")
            print("\nFinal:", (a, b))
            return

        # generate all valid moves
        moves = []

        # Fill Jug1
        if a != jug1:
            moves.append((jug1, b, f"Fill Jug1 -> ({jug1},{b})"))

        # Fill Jug2
        if b != jug2:
            moves.append((a, jug2, f"Fill Jug2 -> ({a},{jug2})"))

        # Empty Jug1
        if a != 0:
            moves.append((0, b, f"Empty Jug1 -> (0,{b})"))

        # Empty Jug2
        if b != 0:
            moves.append((a, 0, f"Empty Jug2 -> ({a},0)"))

        # Pour Jug1 -> Jug2
        transfer = min(a, jug2 - b)
        if transfer > 0:
            newA = a - transfer
            newB = b + transfer
            moves.append((newA, newB,
                          f"Pour Jug1->Jug2 ({a},{b}) -> ({newA},{newB})"))

        # Pour Jug2 -> Jug1
        transfer = min(b, jug1 - a)
        if transfer > 0:
            newA = a + transfer
            newB = b - transfer
            moves.append((newA, newB,
                          f"Pour Jug2->Jug1 ({a},{b}) -> ({newA},{newB})"))

        # add moves to queue
        for newA, newB, action in moves:
            if (newA, newB) not in visited:
                q.append((newA, newB, path + [action]))

    print("No solution")

print("\nBFS with Memorization Steps:")
bfs_with_memo()
```
